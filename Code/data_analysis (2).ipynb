```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt


plt.style.use('classic')
random.seed(42)
```

```python
R_train = pd.read_csv("https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/Lareina/main/Data/Regression_Train%20(2).csv")
R_test = pd.read_csv("https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/Lareina/main/Data/Regression_Test%20(2).csv")
```

```python
R_train.isna().sum()

R_train=R_train.dropna()

R_test.isna().sum()

R_train.head()

R_test.head()
```

```python
from sklearn.metrics import r2_score
```

```python
R_train.head()
```

```python
X_train = pd.DataFrame(R_train['sugar_world_past_ma10'])
y_train = pd.DataFrame(R_train['sugar_world'])
R_test.head()
```

```python
X_test = pd.DataFrame(R_test['sugar_world_past_ma10'])
y_test = pd.DataFrame(R_test['sugar_world'])
```

```python
from sklearn import linear_model

mdl = linear_model.LinearRegression().fit(X_train,y_train)

r2 = r2_score(y_test, mdl.predict(X_test))
print(f'R2 Score: {r2}')

y_pred = pd.DataFrame(np.array(mdl.predict(X_test)).astype(float),columns=['y_pred'])
y_pred
```

```python
fig, ax = plt.subplots()

y_test.hist(ax=ax, label='true', alpha=.75)
y_pred.hist(ax=ax, label='pred', alpha=.75)

ax.legend()
fig.show()


```

```python
import matplotlib.pyplot as plt

# Assuming y_test and y_pred are your true and predicted values

# Create an array of indices for the x-axis
indices = range(len(y_test))

# Plotting the true values as a line
plt.plot(indices, y_test, label='True', marker='o')

# Plotting the predicted values as a line
plt.plot(indices, y_pred, label='Predicted', marker='o')

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('True vs Predicted Line Plot')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

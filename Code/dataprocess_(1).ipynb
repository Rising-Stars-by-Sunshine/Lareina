```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/Lareina/main/Data/sugarprice.csv')
df.head()
```

```python
df.dtypes
```

```python
df['sugar_world'] = pd.to_numeric(df['sugar_world'])
df.dtypes
```

```python
df['sugar_world_past'] =df['sugar_world'].shift(1)
df.head()
```

```python
window = 12
df['sugar_world_past_ma12']=df['sugar_world_past'].rolling(window=window,min_periods=1).mean()
df.head(24)
```

```python
from sklearn.model_selection import TimeSeriesSplit
tss = TimeSeriesSplit()
print(tss)
```

```python
tss = TimeSeriesSplit(gap=0, max_train_size=None, n_splits=2, test_size=None)
```

```python
for train_idx, test_idx in tss.split(df):
    print("TRAIN:", train_idx, "TEST:", test_idx)
```

```python
train_idx
test_idx
train_df = df.filter(items=train_idx, axis=0)
test_df = df.filter(items=test_idx, axis=0)
train_df.head()
```

```python

test_df.head()
```

```python
cols_R = ['sugar_world','sugar_world_past_ma12']
```

```python
df_R_train = train_df[cols_R]
df_R_test = test_df[cols_R]
```

```python
df_R_train.head()
```

```python
df_R_train.to_csv('Regression_Train.csv')
df_R_test.head()
```
